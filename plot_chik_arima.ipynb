```python
import numpy as np
import pandas as pd
from scipy.special import inv_boxcox 
import matplotlib.pyplot as plt
```

```python
df_ = pd.read_csv('data/chik_GO.csv.gz')

df_.head()
```

```python
df_arima = pd.read_csv('forecast_tables/for_gp_chik_5_GO.csv.gz')
df_arima.head()
```

```python
_, ax = plt.subplots()

df_t = df_.tail(12)
df_t['casos'] = inv_boxcox(df_t['casos'].values, 0.05) -1
ax.plot(df_t.date, df_t.casos, color = 'black')
ax.plot(df_arima.date, df_arima.pred, color = 'red')
#ax.fill_between(df_arima.date, df_arima.lower, df_arima.upper, color = 'red', alpha = 0.1)

ax.grid()

plt.show()
```
