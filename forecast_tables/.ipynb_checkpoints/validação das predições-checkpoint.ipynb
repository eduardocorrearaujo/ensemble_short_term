```python
def get_for(week, model = 'ensemble'):

    if model == 'ensemble': 
        try:
            df = pd.read_csv(f'for_ensemble_{for_week}.csv')
        except: 
            df = pd.read_csv(f'for_ensemble_2025_{for_week}.csv')
        
        df = df.rename(columns = {'lower_95':'lower', 'upper_95': 'upper'})

    else: 
        df = pd.DataFrame()
        for state in states_BR: 
            
            df_ = pd.read_csv(f'for_{model}_se_{week}_{state}.csv.gz')
            df_['state'] = state
            df_['model'] = model
            df_ = df_.rename(columns = {'lower_95':'lower', 'upper_95': 'upper'})
            df  = pd.concat([df,df_])

    df.date = pd.to_datetime(df.date)

    df['step'] = [1,2,3]*(28)

    return df 
```

```python
df_for = pd.DataFrame()
df_models = pd.DataFrame()
for for_week in np.arange(min_week, max_week):

    df = get_for(for_week)

    df_for = pd.concat([df_for, df], ignore_index = True)

    df_arima = get_for(for_week, model ='arima')
    df_gp = get_for(for_week, model = 'gp')
    df_lstm = get_for(for_week, model = 'lstm')

    df_models = pd.concat([df_models, df_arima, df_gp, df_lstm], ignore_index = True)

df_for.head()
```

```python
def make_plot(axs, state, model, color, df_data_f, df_for):

    df_for = df_for.copy()
    for ax, step in zip(axs, [1,2,3]):
        
        if model == 'ensemble':
            df_for_step = df_for.loc[(df_for.state == state) & (df_for.step == step)]
        else: 
            df_for_step = df_for.loc[(df_for.model == model) & (df_for.state == state) & (df_for.step == step)]
    
        data_val = df_data_f.loc[(df_data_f.uf == state) & (df_data_f.date.isin(df_for_step.date))]
    
        ax.plot(data_val.date, data_val.casos, color = 'black', label = 'Casos')
    
        ax.plot(df_for_step.date, df_for_step.pred, color = color, label = 'Forecast')
        
        ax.fill_between(df_for_step.date, df_for_step.lower, df_for_step.upper, color = color, alpha =0.2)
        
        ax.set_title(f'{state} - {model} - step: {step}')
        
        ax.legend()
    
        ax.grid()
    
        for label in ax.get_xticklabels():
            label.set_rotation(45)
        
```

Cálculo dos scores:

```python
METRIC = 'LOG_SCORE'
```

```python
def compute_score(state, step, df_data_f, df_for,  df_models, METRIC): 

    df_for = df_for.copy()
    df_models = df_models.copy()

    df_for_step = df_for.loc[(df_for.state == state) & (df_for.step == step)].rename(columns = {'lower': 'lower_95',
                                                            'upper':'upper_95'})
    
    df_models_step = df_models.loc[(df_models.state == state) & (df_models.step == step)].rename(columns = {'lower': 'lower_95',
                                                            'upper':'upper_95'})
    
    data_val = df_data_f.loc[(df_data_f.uf == state) & (df_data_f.date.isin(df_for_step.date))]
        
    scores=[]

    for m in ['arima', 'gp', 'lstm']:
        score_model = Scorer(api_key,
                      df_true = data_val,
                      pred=df_models_step.loc[df_models_step.model == m],
                      dist='log_normal',
                      fn_loss='median',
                      conf_level=0.95)
        
        if METRIC == 'RMSE':
            scores.append(pd.DataFrame([np.sqrt(score_model.mse['pred'])], columns = [f'{m}']))
    
        if METRIC == 'LOG_SCORE':
            scores.append(pd.DataFrame([score_model.log_score[1]['pred']], columns = [f'{m}']))
    
        if METRIC == 'CRPS':
            scores.append(pd.DataFrame([score_model.crps[1]['pred']], columns = [f'{m}']))
    
    for df_, name in zip([df_for_step], ['ensemble']): 
    
        score_model = Scorer(api_key,
                      df_true = data_val,
                      pred=df_for_step,
                      dist='log_normal',
                      fn_loss='median',
                      conf_level=0.95)

        if METRIC == 'RMSE':
            scores.append(pd.DataFrame([np.sqrt(score_model.mse['pred'])], columns = [f'{name}']))

        if METRIC == 'LOG_SCORE':
            scores.append(pd.DataFrame([score_model.log_score[1]['pred']], columns = [f'{name}']))
    
        if METRIC == 'CRPS':
            scores.append(pd.DataFrame([score_model.crps[1]['pred']], columns = [f'{name}']))
    
    df_scores_metric = pd.concat(scores, axis =1).dropna()

    df_scores_metric = df_scores_metric.T.rename(columns = {0:METRIC})
    
    return df_scores_metric
```

```python
state = 'MG'
step = 1 
METRIC = 'RMSE'
```

```python
df_score_end = pd.DataFrame()

for state in states_BR:
    if state!='BR':
        for step in [1,2,3]:
    
            df_score = pd.concat([compute_score(state, step, df_data_f, df_for,  df_models, 'RMSE'),
              compute_score(state, step, df_data_f, df_for,  df_models, 'LOG_SCORE'),
              compute_score(state, step, df_data_f, df_for,  df_models, 'CRPS')], axis = 1)
            df_score['state'] = state
            df_score['step'] = step 
            df_score.index.name = 'model'
            df_score = df_score.reset_index()
    
        df_score_end = pd.concat([df_score_end, df_score])
```

```python
df_score_end.to_csv('scores_models.csv', index = False)
```
