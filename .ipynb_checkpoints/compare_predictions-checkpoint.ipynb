```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.special import inv_boxcox 
# Definir a cor das bordas (spines) como cinza
mpl.rcParams['axes.edgecolor'] = 'gray'

# Definir a cor das linhas dos ticks maiores e menores como cinza
mpl.rcParams['xtick.color'] = 'gray'
mpl.rcParams['ytick.color'] = 'gray'
mpl.rcParams['xtick.labelcolor'] = 'black'
mpl.rcParams['ytick.labelcolor'] = 'black'
```

```python
states_ne = ['AL', 'BA', 'CE','MA', 'PB', 'PE', 'PI', 'SE', 'RN']
states_se = ['SP', 'MG', 'RJ', 'ES']
states_sul = ['RS', 'SC', 'PR']
states_ce = ['DF', 'GO', 'MT', 'MS']
states_no = ['AM', 'AP', 'TO', 'RR', 'RO' ,'AC', 'PA']

states_BR = states_ne+states_se+states_no+states_ce+states_sul
```

```python
df = pd.read_csv('./data/dengue_update.csv.gz', index_col ='Unnamed: 0')
df.date = pd.to_datetime(df.date)
df.drop(['regional_geocode'], axis =1, inplace= True)

df = df.groupby(['date', 'uf']).sum().reset_index()

df = df.loc[df.date >= '2024-06-01']
#df = df.loc[df.uf == 'PR']
df.tail()
```

```python
df_for_ens = pd.read_csv('forecast_tables/for_ensemble_40.csv')

df_for_ens.date = pd.to_datetime(df_for_ens.date)

df_for_ens.head()
```

```python
df_for_ens1 = pd.read_csv('forecast_tables/for_ensemble_41.csv')

df_for_ens1.date = pd.to_datetime(df_for_ens1.date)

df_for_ens1.head()
```

```python
df_for_ens2 = pd.read_csv('forecast_tables/for_ensemble_42.csv')

df_for_ens2.date = pd.to_datetime(df_for_ens2.date)

df_for_ens2.head()
```

```python
for state in states_BR: 
    df_ = df.loc[df.uf == state].sort_values(by = 'date')
    
    df_for_ens_ = df_for_ens.loc[df_for_ens.state == state]

    df_for_ens1_ = df_for_ens1.loc[df_for_ens1.state == state]

    df_for_ens2_ = df_for_ens2.loc[df_for_ens2.state == state]
    
    fig,ax = plt.subplots(1,1, figsize = (8, 5))
        
    ax.plot(df_.date,  df_.casos, color = 'black', linewidth = 2, label = 'Casos')
    
    ax.plot(df_for_ens_.date,  df_for_ens_.pred, color = 'tab:orange', label = 'Forecast SE 40')
                    
    ax.fill_between(df_for_ens_.date, df_for_ens_.lower, df_for_ens_.upper, color = 'tab:orange', alpha = 0.1)

    ax.plot(df_for_ens1_.date,  df_for_ens1_.pred, color = 'tab:red', label = 'Forecast SE 41')
                    
    ax.fill_between(df_for_ens1_.date, df_for_ens1_.lower, df_for_ens1_.upper, color = 'tab:red', alpha = 0.1)

    ax.plot(df_for_ens2_.date,  df_for_ens2_.pred, color = 'tab:purple', label = 'Forecast SE 42')
                    
    ax.fill_between(df_for_ens2_.date, df_for_ens2_.lower, df_for_ens2_.upper, color = 'tab:purple', alpha = 0.1)

    ax.grid()
    ax.legend()
    ax.set_title(f'Validação forecast - {state}')
    plt.savefig(f'./figures/val_{state}.png', dpi = 300, bbox_inches = 'tight')
    plt.show()
```
