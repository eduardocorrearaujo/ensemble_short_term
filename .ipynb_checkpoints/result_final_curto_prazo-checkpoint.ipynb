```python
import numpy as np
import pandas as pd
from epiweeks import Week
import matplotlib.pyplot as plt 
from mosqlient.models.score import Scorer
from datetime import datetime, timedelta
from scipy.special import inv_boxcox 
import matplotlib as mpl
from bayes_ensemble import *
# Definir a cor das bordas (spines) como cinza
mpl.rcParams['axes.edgecolor'] = 'gray'

# Definir a cor das linhas dos ticks maiores e menores como cinza
mpl.rcParams['xtick.color'] = 'gray'
mpl.rcParams['ytick.color'] = 'gray'
mpl.rcParams['xtick.labelcolor'] = 'black'
mpl.rcParams['ytick.labelcolor'] = 'black'
```

```python
states_ne = ['AL', 'BA', 'CE','MA', 'PB', 'PE', 'PI', 'SE', 'RN']
states_se = ['SP', 'MG', 'RJ', 'ES']
states_sul = ['RS', 'SC', 'PR']
states_ce = ['DF', 'GO', 'MT', 'MS']
states_no = ['AM', 'AP', 'TO', 'RR', 'RO' ,'AC', 'PA']

states_ = states_ne+states_se+states_no+states_ce+states_sul
states_BR = states_ne+states_se+states_no+states_ce+states_sul+['BR']
```

```python
df = pd.read_csv('data/dengue_BR.csv.gz')
df.date = pd.to_datetime(df.date)
df.head()
```

```python
max_date = df.date.max()

for_week = Week.fromdate(df.date.max()).week
for_week
```

```python
def get_preds(state, model, week): 

    try: 
        df = pd.read_csv(f'forecast_tables/for_{model}_se_{week}_{state}.csv.gz')
        
    except: 
        df = pd.read_csv(f'forecast_tables/for_{model}_{week}_{state}.csv',usecols = ['date', 'lower_2_5', 'forecast', 'upper_97_5'])

        df = df.rename(columns = {'lower_2_5': 'lower', 'forecast': 'pred', 'upper_97_5': 'upper'})

    df.date = pd.to_datetime(df.date)
        
    return df


def get_all_preds(state, week):

    df_pred1 = get_preds(state, 'lstm', week = week)
    df_pred1['model_id'] = 1
    
    df_pred2 = get_preds(state, 'gp', week = week)
    df_pred2['model_id'] = 2

    df_pred3 = get_preds(state, 'arima', week = week)
    df_pred3['model_id'] = 3

    return pd.concat([df_pred1, df_pred2, df_pred3])


def get_ensemble(state, max_date, week):

    df = pd.read_csv(f'data/dengue_{state}.csv.gz')
    df.date = pd.to_datetime(df.date)
    df = df.loc[df.date == max_date]
    
    obs = inv_boxcox(df['casos'].values[0], 0.05) -1 

    df_preds = get_all_preds(state= state, week= for_week-1)

    df_preds = df_preds.loc[df_preds.date == max_date]

    df_preds_ =  get_df_log_pars(df_preds[['pred', 'lower', 'upper', 'model_id']].reset_index(drop = True), alpha = 0.95)

    output = find_opt_CRPS_weights(obs = obs, 
                    ms = df_preds_.mu, 
                    vs = df_preds_.sigma**2)

    df_for = get_all_preds(state= state, week= for_week)

    preds_ens_crps = np.empty((0, 3))

    p = np.array([0.5, 0.05, 0.95])
    
    for date in df_for.date.unique():   
    
        weights_crps = output['weights']
    
        preds = df_for.loc[df_for.date == date][['pred', 'lower', 'upper', 'model_id']].reset_index(drop = True)
    
        preds = preds.sort_values(by = 'model_id')
            
        preds=  get_df_log_pars(preds, alpha = 0.95)
    
        LS_pool_crps = pool_par_gauss(alpha = weights_crps, m = preds.mu,
                v = preds.sigma**2)
    
        quantiles_crps = lognorm.ppf(p, s=LS_pool_crps[1], scale=np.exp(LS_pool_crps[0]))
    
        preds_ens_crps = np.concatenate((preds_ens_crps, quantiles_crps.reshape(1, 3)))

    df_ens_crps = pd.DataFrame(preds_ens_crps, columns = ['pred', 'lower', 'upper'])

    df_ens_crps['date'] =  df_for.date.unique()
    
    df_ens_crps.date = pd.to_datetime(df_ens_crps.date)

    df_ens_crps['state'] = state
    
    return df_ens_crps, weights_crps 
```

```python
df_crps = pd.DataFrame()
df_w = pd.DataFrame()
for state in states_BR:
    df_crps_ , weights_ = get_ensemble(state, max_date, for_week)
    df_crps = pd.concat([df_crps, df_crps_], ignore_index = True)

    df_w_ = pd.DataFrame(weights_, columns = ['weights'])
    df_w_['model']  = ['model_1', 'model_2', 'model_3']
    df_w_['state'] = state
    df_w = pd.concat([df_w, df_w_])

df_crps.to_csv(f'forecast_tables/for_ensemble_{for_week}.csv', index = False)
df_crps.head()
```

### Making the plot: 

```python
df_ens_ = pd.read_csv('ensemble_bayes_2025.csv.gz')
df_ens_.date = pd.to_datetime(df_ens_.date)
#df_ens.set_index('date', inplace = True)
df_ens_['epiweek'] = [str(Week.fromdate(x)) for x in df_ens_.date]
df_ens_ = df_ens_.rename(columns = {'mean_ensemble_23': 'mu_ensemble_23', 'mean_ensemble_24': 'mu_ensemble_24'})

df_ens_ = df_ens_.loc[df_ens_.date <= df_crps.date.max()]

df_ens_.head()
```

```python
from matplotlib.ticker import MaxNLocator

for state in states_: 

    try:
    
        data = pd.read_csv(f'data/dengue_{state}.csv.gz')
        
        data.date = pd.to_datetime(data.date)
    
        data = data.loc[data.date >= '2024-08-01']
    
        data['casos'] = inv_boxcox(data['casos'].values, 0.05) -1
    
        df_ens_crps = df_crps.loc[df_crps.state == state]
        #df_ens = df_ens_.loc[df_ens_.state == state].reset_index(drop=True)
        df_ens = df_ens_.loc[df_ens_.state == state].reset_index(drop=True)
        
        fig,ax = plt.subplots(1,2, figsize = (13, 5))
    
        for ax_ in ax.ravel(): 
            ax_.plot(data.date,  data.casos, color = 'black', linewidth = 2, linestyle = '-', label = 'Casos')
        
            ax_.plot([data.date.values[-1], df_ens_crps.date.values[0]], [data['casos'].values[-1], df_ens_crps.pred.values[0]], ls = '--', color = 'black')
        
            ax_.plot(df_ens_crps.date,  df_ens_crps.pred, color = 'tab:red', label = 'Forecast curto prazo')
                
            ax_.fill_between(df_ens_crps.date, df_ens_crps.lower, df_ens_crps.upper, color = 'tab:red', alpha = 0.1)
    
            ax_.plot(df_ens.date, df_ens.pred_ensemble_23, color = '#35DB65', label = 'Ensemble (2023)')
            
            ax_.fill_between(df_ens.date, df_ens.lower_ensemble_23, df_ens.upper_ensemble_23, color = '#35DB65', alpha = 0.2)
            
            ax_.plot(df_ens.date, df_ens.pred_ensemble_24, color = '#3577DB', label = 'Ensemble (2024)')
                
            ax_.fill_between(df_ens.date, df_ens.lower_ensemble_24, df_ens.upper_ensemble_24, color = '#3577DB', alpha = 0.2)
        
    
        for ax_ in ax.ravel():
            ax_.set_xlabel('Data')
            ax_.set_ylabel('Novos Casos')
            ax_.set_title(f'Forecast casos prováveis - {state}')
            ax_.legend()
            ax_.grid()
    
        ax[0].set_ylim(min(0.9*min(data.casos), (min(df_ens_crps.lower))), max(1.25*max(df_ens_crps.upper), max(data.casos)))
    
        fig.autofmt_xdate(rotation=30, ha='center')
        
    
        plt.savefig(f'./figures/forecast_{for_week}_{state}.png', dpi = 300, bbox_inches = 'tight')

    except: 
        print(state)
plt.show()
```
