```python
import numpy as np
import pandas as pd 
from epiweeks import Week
from scipy.stats import boxcox
import matplotlib.pyplot as plt
```

```python
def calcular_metricas_por_janela(array, tamanho_janela, funcoes):
    # Criar um array com as janelas deslizantes
    janelas = np.lib.stride_tricks.sliding_window_view(array, tamanho_janela)

    # Aplicar as funções de interesse em cada janela
    resultados = [func(janela, axis=0) for func in funcoes for janela in janelas]
    
    
    return np.array(resultados)
```

```python
def get_slope(casos, axis = 0): 
     
    return np.polyfit(np.arange(0,4), casos, 1)[0]
```

```python
def get_clima(state): 

    df_climate = pd.read_parquet(f'../../paper-dengue-sc/data/climate/{state}_climate.parquet', 
                            columns = ['geocodigo', 'temp_med', 'umid_med', 'precip_tot'])
    df_climate = df_climate.rename(columns = {'umid_med': 'rel_humid_med'})
    
    df_climate = df_climate.reset_index().merge(map_geo_reg, left_on = 'geocodigo', right_on = 'geocode').drop(['geocodigo', 'geocode'], axis =1)
    
    df_climate.date = pd.to_datetime(df_climate.date)
    
    df_climate.set_index('date', inplace = True)
    
    df_climate = df_climate.groupby('regional_geocode').resample('W-SUN').agg({'temp_med': 'mean',
                                                   'rel_humid_med': 'mean',
                                                   'precip_tot': 'sum'}).reset_index()

    df_temp = df_climate.pivot(index = 'date', columns = 'regional_geocode', values = 'temp_med')

    df_temp.columns = 'temp_' + df_temp.columns.astype(str)

    df_hum = df_climate.pivot(index = 'date', columns = 'regional_geocode', values = 'rel_humid_med')

    df_hum.columns = 'rel_hum_' + df_hum.columns.astype(str)

    df_prec = df_climate.pivot(index = 'date', columns = 'regional_geocode', values = 'precip_tot')

    df_prec.columns = 'precip_' + df_prec.columns.astype(str)

    df_end = df_temp.merge(df_hum, left_index = True, right_index = True).merge(df_prec, left_index = True, right_index = True)

    return df_end
```

```python
map_geo_reg = pd.read_csv('map_geo_reg.csv', index_col = 'Unnamed: 0')
```

### Aggregate data using gtrends (only last five years available weekly): 

```python
state = 'PR'
```

```python
gt = pd.read_csv('/Users/eduardoaraujo/Downloads/PR_gtrends.csv', skiprows =1)

gt.columns = ['date', 'trends']

gt.date = pd.to_datetime(gt.date)


gt['diff_trends'] = np.concatenate( (np.array([np.nan]), np.diff(gt['trends'], 1)))
    
array = np.array(gt.trends)
tamanho_janela = 4
    
gt['trends_mean'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [np.mean])))
    
    
gt['trends_std'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [np.std])))
    
gt['trends_slope'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [get_slope])))

gt = gt.dropna()


gt.set_index('date', inplace = True)

gt.head()
```

```python
df = df_all.loc[df_all.uf == state].copy()

df.date = pd.to_datetime(df.date)
df.set_index('date', inplace = True)

df = df[['casos']].resample('W-SUN').sum()

df['casos'] = boxcox(df.casos+1, lmbda=0.05)

df['SE'] = [Week.fromdate(x) for x in df.index]

df = df[['SE', 'casos']].sort_index()
    
    
df['SE'] = df['SE'].astype(str).str[-2:].astype(int)
    
df['SE'] = df['SE'].replace(53,52)
    
df['diff_casos'] = np.concatenate( (np.array([np.nan]), np.diff(df['casos'], 1)))
    
array = np.array(df.casos)
tamanho_janela = 4
    
df['casos_mean'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [np.mean])))
    
    
df['casos_std'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [np.std])))
    
df['casos_slope'] =  np.concatenate( (np.array([np.nan, np.nan, np.nan]), calcular_metricas_por_janela(array, tamanho_janela, [get_slope])))
    
df = df.dropna()

df.head()
```
